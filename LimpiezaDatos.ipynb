## Detalles hoteles

```python
hoteles = pd.read_csv('data/detalles.csv')
```

```python
hoteles
```

```python
hoteles['provincia'] = hoteles['ubicacion'].apply(lambda x: x.split(",")[0])
hoteles['ubicacion'] = hoteles['ubicacion'].apply(lambda x: x.split(",")[1])
```

```python
hoteles
```

```python
len(hoteles['provincia'].unique())
```

```python
import sqlite3
connection = sqlite3.connect("proyecto.db")
connection.text_factory = str
hoteles.to_sql(name='hoteles', con=connection, if_exists = 'append', index=False)
```

```python
ax = hoteles.groupby(['provincia']).size().sort_values(ascending=False)[:20].plot.bar(title = u'Mayor cantidad de hoteles según provincia', legend=True)

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
```
